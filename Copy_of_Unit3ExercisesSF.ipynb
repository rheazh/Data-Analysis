## The First Four Moments: Mean, Variance, Skew, and Kurtosis

When we talk about a probablity dist, we often care about at four things:
- The expected value, or where a distribution is centered (mean)
- The amount of variation from that centered value (variation)
- Whether or not our distribution favors, leans towards, or skews towards certain values. (skew)
- How often to expect unusually high or low values. (kurtosis)

**Task4**:

Plot the pdfs of each of the following distributions: $Normal(μ=0,σ=1)$, $StudentT(ν=1,μ=0,σ=1)$, $Exponential(λ=1/2)$, $Gamma(α=2,β=1)$, $HalfStudentT(ν=1,σ1)$

For the T distribution, use

```
support=(-4,4)
```
inside the plot_pdf function.


````python
import matplotlib.pyplot as plt
from scipy.stats import norm, t, expon, gamma, halfcauchy


x_norm = np.linspace(-4, 4, 1000)
x_t = np.linspace(-4, 4, 1000)
x_exp = np.linspace(0, 10, 1000)
x_gamma = np.linspace(0, 10, 1000)
x_half_t = np.linspace(0, 10, 1000)


# Normal Distribution
pdf_norm = norm.pdf(x_norm, loc=0, scale=1)


# Student's T Distribution
pdf_t = t.pdf(x_t, df=1, loc=0, scale=1)


# Exponential Distribution
pdf_exp = expon.pdf(x_exp, scale=2)  # λ = 1/2, scale = 1/λ


# Gamma Distribution
pdf_gamma = gamma.pdf(x_gamma, a=2, scale=1)


# Half Student's T Distribution
pdf_half_t = 2 * halfcauchy.pdf(x_half_t, scale=1)  # scale corresponds to σ1


plt.figure(figsize=(12, 8))


# Normal Distribution
plt.subplot(3, 2, 1)
plt.plot(x_norm, pdf_norm, label='Normal(μ=0, σ=1)')
plt.title('Normal Distribution')
plt.xlim(-4, 4)
plt.ylim(0, 0.5)




# Student's T Distribution
plt.subplot(3, 2, 2)
plt.plot(x_t, pdf_t, label="Student's T(ν=1, μ=0, σ=1)")
plt.title("Student's T Distribution")
plt.xlim(-4, 4)
plt.ylim(0, 0.5)




# Exponential Distribution
plt.subplot(3, 2, 3)
plt.plot(x_exp, pdf_exp, label='Exponential(λ=1/2)')
plt.title('Exponential Distribution')
plt.xlim(0, 10)
plt.ylim(0, 0.5)




# Gamma Distribution
plt.subplot(3, 2, 4)
plt.plot(x_gamma, pdf_gamma, label='Gamma(α=2, β=1)')
plt.title('Gamma Distribution')
plt.xlim(0, 10)
plt.ylim(0, 0.5)




# Half Student's T Distribution
plt.subplot(3, 2, 5)
plt.plot(x_half_t, pdf_half_t, label='Half Student\'s T(ν=1, σ1=1)')
plt.title("Half Student's T Distribution")
plt.xlim(0, 10)
plt.ylim(0, 0.5)


plt.tight_layout()
plt.show()
````

**Task9**:

Plot the pdfs (techincally the pmfs) of the following distributions: $Bernoulli(p=0.7)$, $Binomial(n=10,p=0.7)$, $Poisson(λ=7)$, $Geometric(p=0.7)$, $HyperGeometric(N=50,k=35,n=15)$, $Categorical(p=[0.05,0.3,0.05,0.1,0.2,0.4])$

Additionally, compute the first four moments for all those distributions--except the Categorical distribution.

````python
from scipy.stats import bernoulli
import seaborn as sb
p_bernoulli = 0.7

data_bern = bernoulli.rvs(size=1000,p=0.7)
ax = sb.distplot(data_bern,)
kde=True,
ax.set(xlabel='Bernouli', ylabel='Frequency')
mean_bernoulli = p_bernoulli
var_bernoulli = p_bernoulli * (1 - p_bernoulli)
skew_bernoulli = (1 - 2 * p_bernoulli) / np.sqrt(p_bernoulli * (1 - p_bernoulli))
kurt_bernoulli = 3 - 6 * p_bernoulli + 6 * p_bernoulli**2
print(mean_bernoulli)
print(var_bernoulli)
print(skew_bernoulli)
print(kurt_bernoulli)
````

````python
from scipy.stats import binom
import matplotlib.pyplot as plt
n_binomial = 10
p_binomial = 0.7
# setting the values
# of n and p
n = 10
p = 0.7
# defining list of r values
r_values = list(range(n + 1))
# list of pmf values
dist = [binom.pmf(r, n, p) for r in r_values ]
# plotting the graph
plt.bar(r_values, dist)
plt.show()
mean_binomial = n_binomial * p_binomial
var_binomial = n_binomial * p_binomial * (1 - p_binomial)
skew_binomial = (1 - 2 * p_binomial) / np.sqrt(n_binomial * p_binomial * (1 - p_binomial))
kurt_binomial = (1 - 6 * p_binomial * (1 - p_binomial)) / (n_binomial * p_binomial * (1 - p_binomial)) + 3
print(mean_binomial)
print(var_binomial)
print(skew_binomial)
print(kurt_binomial)
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom, poisson, geom, hypergeom

lambda_poisson = 7

x_poisson = np.arange(0, 20)
pmf_poisson = poisson.pmf(x_poisson, lambda_poisson)
plt.subplot(2, 3, 3)
plt.bar(x_poisson, pmf_poisson, alpha=0.7)
plt.title('Poisson PMF (λ=7)')
mean_poisson = lambda_poisson
var_poisson = lambda_poisson
print(f"Poisson: Mean={mean_poisson}, Variance={var_poisson}")
skew_poisson = 1 / np.sqrt(lambda_poisson)
kurt_poisson = 1 / lambda_poisson + 3
print(skew_poisson)
print(kurt_poisson)


````

````python
p_geometric = 0.7
x_geometric = np.arange(1, 20)
pmf_geometric = geom.pmf(x_geometric, p_geometric)
plt.subplot(2, 3, 4)
plt.bar(x_geometric, pmf_geometric, alpha=0.7)
plt.title('Geometric PMF (p=0.7)')
mean_geometric = 1 / p_geometric
var_geometric = (1 - p_geometric) / p_geometric**2
print(f"Geometric: Mean={mean_geometric}, Variance={var_geometric}")
skew_geometric = (2 - p_geometric) / np.sqrt(1 - p_geometric)
kurt_geometric = (6 - 6 * p_geometric) / (1 - p_geometric) + 3
print(skew_geometric)
print(kurt_geometric)
````

````python
N_hyper = 50
k_hyper = 35
n_hyper = 15
x_hyper = np.arange(0, min(k_hyper, n_hyper) + 1)
pmf_hyper = hypergeom.pmf(x_hyper, N_hyper, k_hyper, n_hyper)
plt.subplot(2, 3, 5)
plt.bar(x_hyper, pmf_hyper, alpha=0.7)
plt.title('Hypergeometric PMF (N=50, k=35, n=15)')
mean_hyper = (n_hyper * k_hyper) / N_hyper
var_hyper = (n_hyper * k_hyper * (N_hyper - k_hyper) * (N_hyper - n_hyper)) / (N_hyper**2 * (N_hyper - 1))
print(f"Hypergeometric: Mean={mean_hyper}, Variance={var_hyper}")
skew_hyper = np.nan
kurt_hyper = np.nan
print(skew_hyper)
print(kurt_hyper)
````
